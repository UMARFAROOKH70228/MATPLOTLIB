```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
df = iris.drop(['species'],axis=1)
```

```python
df
```

```python
df.plot(kind='area')
```

```python
ax=df.plot.scatter(x='sepal_length',y='petal_length',label='length')
df.plot.scatter(x='sepal_length',y='petal_width',label='width',ax=ax,color='r')
```
